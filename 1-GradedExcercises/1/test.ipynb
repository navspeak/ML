```python
import pandas as pd

df = pd.read_excel("college.xlsx")

df = df.drop(columns=["Unnamed: 0"])
df.head()
```

```python
df.describe().round(2)
```

```python
df_all_numeric = pd.get_dummies(df, columns=["Private"])
df_all_numeric.head(1)
```

```python
df_all_numeric = pd.get_dummies(df, columns=["Private"], drop_first=True)
df_all_numeric.head(1)
```

```python
X = df_all_numeric.drop(columns=["Grad.Rate"])
y = df["Grad.Rate"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
coefficients = pd.Series(model.coef_, index=X.columns)
print(coefficients.round(2))
model.coef_
print(model.intercept_)
```

# Predictions

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

# Evaluation Metrics

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
train_r2 = round(r2_score(y_train, y_train_pred), 2)
train_mse = round(mean_squared_error(y_train, y_train_pred), 2)
train_mae = round(mean_absolute_error(y_train, y_train_pred), 2)

print("Training Metrics")
print("R2:", train_r2)
print("MSE:", train_mse)
print("MAE:", train_mae)
```

# Testing Metrics

1. Actual  `y = [10, 20, 30]`
2. Predicted  `ŷ = [12, 18, 33]`
3. Error `[2, -2, 3]`
4. SSE `2^2+(−2)^2+3^2=4+4+9=17`
5. MSE `SSE/n = 17/3 = 5.67`
6. MAE `(|2| + |3| + |3|) = 7/3 = 2.33`


```python
test_r2 = round(r2_score(y_test, y_test_pred), 2)
test_mse = round(mean_squared_error(y_test, y_test_pred), 2)
test_mae = round(mean_absolute_error(y_test, y_test_pred), 2)

print("Testing Metrics")
print("R2:", test_r2)
print("MSE:", test_mse) # = SSE / n 
print("MAE:", test_mae)
```

## Applications of Similarity Metrics

### Setup and Data Preprocessing

```python
import warnings; warnings.filterwarnings('ignore'); import sys; import os; sys.stderr = open(os.devnull, 'w')  # Suppress warnings
import numpy as np; import pandas as pd; import matplotlib.pyplot as plt; import seaborn as sns  # Data processing and visualisation
from sklearn.preprocessing import MinMaxScaler  # Feature scaling
from sklearn.neighbors import KNeighborsRegressor  # KNN regression
from sklearn.metrics import mean_squared_error  # MSE
from scipy.cluster.hierarchy import linkage, dendrogram  # Hierarchical clustering
```

```python
df = pd.read_csv('carprice.csv', index_col = 'car_ID'); df.sample(3)  # Data loading
```

The data has no missing values

```python
df.shape  # Data size
```

```python
df.describe().T  # Studying the numerical data
```

```python
temp = df.corr(); temp[np.abs(temp) < 0.5]  # Correlation analysis
```

Retain a subset of numerical feature to work with

```python
cols = ['boreratio', 'horsepower', 'price']
```

```python
df = df[cols]; df.sample(3)  # Subsetting the dataset
```

We need to rescale all numerical features to the same scale

```python
scaler = MinMaxScaler(); df_s = scaler.fit_transform(df); df_s  # Rescaling numerical features
```

```python
df_s = pd.DataFrame(df_s, columns = cols, index = df.index); df_s.sample(3)  # Scaled dataset
```

### Supervised ML / KNN Regressor

```python
X = df_s.drop(columns = 'price'); y = df_s['price']  # Input and output
```

```python
knn = KNeighborsRegressor(); knn.fit(X, y)  # Default K is 5 and default metric is Minkowski with p = 2
y_pred = knn.predict(X); mse = mean_squared_error(y, y_pred); print('Training MSE = {}'.format(np.round(mse, 4)))  # Predictions and MSE
```

```python
for k in [2, 5, 10]:  # Different K
    for p_val in [1, 1.5, 2, 5, 100]:  # Different Minkowski p values
        knn = KNeighborsRegressor(n_neighbors = k, p = p_val); knn.fit(X, y)
        y_pred = knn.predict(X); mse = mean_squared_error(y, y_pred)
        print('Training MSE = {} for p = {} and K = {}'.format(np.round(mse, 4), p_val, k))
```

### Unsupervised ML / Hierarchical Clustering

```python
clusters = linkage(y = df_s)  # Default method is single linkage and default metric is Euclidean distance
plt.figure(figsize = (10, 4))
dendrogram(Z = clusters, orientation = 'right', labels = df_s.index)  # Dendrogram from clustering results
plt.title('Dendrogram - Single Linkage - Euclidean Distance'); plt.xlabel('Cluster distance'); plt.tight_layout();
```

```python
clusters = linkage(y = df_s, metric = 'cityblock')  # Manhattan distance
plt.figure(figsize = (10, 4))
dendrogram(Z = clusters, orientation = 'right', labels = df_s.index)  # Dendrogram from clustering results
plt.title('Dendrogram - Single Linkage - Manhattan Distance'); plt.xlabel('Cluster distance'); plt.tight_layout();
```

```python
clusters = linkage(y = df_s, method = 'complete')  # Complete linkage with Euclidean distance
plt.figure(figsize = (10, 4))
dendrogram(Z = clusters, orientation = 'right', labels = df_s.index)  # Dendrogram from clustering results
plt.title('Dendrogram - Complete Linkage - Euclidean Distance'); plt.xlabel('Cluster distance'); plt.tight_layout();
```

```python
clusters = linkage(y = df_s, method = 'complete', metric = 'cityblock')  # Complete linkage with Manhattan distance
plt.figure(figsize = (10, 4))
dendrogram(Z = clusters, orientation = 'right', labels = df_s.index)  # Dendrogram from clustering results
plt.title('Dendrogram - Complete Linkage - Manhattan Distance'); plt.xlabel('Cluster distance'); plt.tight_layout();
```

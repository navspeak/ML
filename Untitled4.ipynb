Load the breast cancer dataset from scikit-learn; you can do this by running load_breast_cancer() from sklearn.datasets

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

The target variable here is whether the target has breast cancer or not; you can obtain this by using the .target attribute on your loaded data

```python
import pandas as pd
X = data.data
y = data.target
df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y
# df.info()
for i, class_name in enumerate(data.target_names):
    print(f'{i}: {class_name}')
    count = (y == i).sum()
    print(f'  {class_name}: {count} ({count/len(y):.1%})')
```

Using sklearn.model_selection.train_test_split(), split the data as follows:

Use an 80:20 train:test split

Set random_state = 9001 so that the results you obtain can be compared to ours

Stratify the target feature labels so we have an equal distribution in both test and training sets (set stratify = y)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=9001, stratify=y)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Build a base random forest model using RandomForestClassifier() from sklearn.ensemble

Set class_weight = 'balanced'

Set random_state = 9001 so that the results you obtain can be compared to ours

Set n_estimators = 100

Set max_depth = 4
  

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, max_depth=4, class_weight='balanced', random_state=9001)
rf.fit(X_train, y_train)
rf.feature_importances_
```

During evaluation, generate

A confusion matrix

A classification report; you can use classification_report() from sklearn.metrics to achieve this

The ROC AUC score; you can use roc_auc_score() from sklearn.metrics to achieve this


```python
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, accuracy_score

y_pred_train = rf.predict(X_train)
y_pred_test = rf.predict(X_test)

print("Confusion Matrix for Training Set:")
print(confusion_matrix(y_train, y_pred_train))

print("\nConfusion Matrix for Test Set:")
print(confusion_matrix(y_test, y_pred_test))

```

```python
train_acc = accuracy_score(y_train, y_pred_train)
test_acc = accuracy_score(y_test, y_pred_test)
print("Accuracy Score for Training Set:", train_acc)
print("Accuracy Score for Test Set:", test_acc)
```

```python
print("Classification Report for Training Set:")
print(classification_report(y_train, y_pred_train))
print("\n Classification Report for Test Set:")
print(classification_report(y_test, y_pred_test))
```

```python
train_prob = rf.predict_proba(X_train)[:, 1]
test_prob = rf.predict_proba(X_test)[:, 1]

train_auc = roc_auc_score(y_train, train_prob)
test_auc = roc_auc_score(y_test, test_prob)

print("ROC AUC Score for Training Set:", train_auc)
print("ROC AUC Score for Test Set:", test_auc)
```

Build a base LGBMClassifier() model

Set class_weight = 'balanced'

Set random_state = 9001 so that the results you obtain can be compared to ours

Set verbose = -1

Set n_estimators = 100

Set max_depth = 2

Set learning_rate = 0.1



```python
from lightgbm import LGBMClassifier
gbt = LGBMClassifier(n_estimators = 100,
                    max_depth = 2,
                    learning_rate = 0.1,
                    class_weight = 'balanced',
                    random_state = 9001,
                    verbose = -1)
gbt.fit(X_train, y_train)
```

During evaluation, generate

A confusion matrix

A classification report; you can use classification_report() from sklearn.metrics to achieve this

The ROC AUC score; you can use roc_auc_score() from sklearn.metrics to achieve this


```python
y_pred_train = gbt.predict(X_train)
y_pred_test = gbt.predict(X_test)

print("Confusion Matrix for Training Set:")
print(confusion_matrix(y_train, y_pred_train))

print("\nConfusion Matrix for Test Set:")
print(confusion_matrix(y_test, y_pred_test))
```

```python
print("Classification Report for Training Set:")
print(classification_report(y_train, y_pred_train))
print("\n Classification Report for Test Set:")
print(classification_report(y_test, y_pred_test))
```

```python
train_acc = accuracy_score(y_train, y_pred_train)
test_acc = accuracy_score(y_test, y_pred_test)
print("Accuracy Score for Training Set:", train_acc)
print("Accuracy Score for Test Set:", test_acc)
```

```python
train_prob = gbt.predict_proba(X_train)[:, 1]
test_prob = gbt.predict_proba(X_test)[:, 1]

train_auc = roc_auc_score(y_train, train_prob)
test_auc = roc_auc_score(y_test, test_prob)

print("ROC AUC Score for Training Set:", train_auc)
print("ROC AUC Score for Test Set:", test_auc)
```

```python
print("\n--- Classification Summary ---\n")
print("Random Forest Model:")
print(f"  Training Accuracy: {accuracy_score(y_train, rf.predict(X_train)):.4f}")
print(f"  Test Accuracy: {accuracy_score(y_test, rf.predict(X_test)):.4f}")
print(f"  Training ROC AUC: {roc_auc_score(y_train, rf.predict_proba(X_train)[:, 1]):.4f}")
print(f"  Test ROC AUC: {roc_auc_score(y_test, rf.predict_proba(X_test)[:, 1]):.4f}")

print("\nLightGBM Model:")
print(f"  Training Accuracy: {accuracy_score(y_train, gbt.predict(X_train)):.4f}")
print(f"  Test Accuracy: {accuracy_score(y_test, gbt.predict(X_test)):.4f}")
print(f"  Training ROC AUC: {roc_auc_score(y_train, gbt.predict_proba(X_train)[:, 1]):.4f}")
print(f"  Test ROC AUC: {roc_auc_score(y_test, gbt.predict_proba(X_test)[:, 1]):.4f}")

print("\nObservations:")
print("- Both models show excellent performance on the training data, with high accuracy and ROC AUC scores.")
print("- On the test set, both models maintain strong performance, indicating good generalization.")
print("- The LightGBM model achieved perfect accuracy and ROC AUC on the training set, suggesting it might be overfitting slightly more than the Random Forest model, although its test performance is still very good and comparable to Random Forest.")
```

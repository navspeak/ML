````python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([1,2,3,4,5]).reshape(-1,1) #matrix (samples x features)
y = np.array([50,55,65,70,80])

model = LinearRegression().fit(X,y)

print(model.predict([[3.5]]))  # interpolation
print(model.predict([[10]]))   # extrapolation

print(model.intercept_)   # β0
print(model.coef_)        # β1
````

## Multicollinearity

````python
import numpy as np
import pandas as pd

# create dataset
np.random.seed(0)

size = np.random.randint(1000, 2000, 10)
rooms = size / 500 + np.random.normal(0,0.1,10)   # highly correlated with size
price = size * 200 + rooms * 5000

df = pd.DataFrame({
    "size": size,
    "rooms": rooms,
    "price": price
})

print(df)
````

## VIF

````python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = df[["size","rooms"]]

vif = pd.DataFrame()
vif["feature"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif)
print(X.corr())
````

- In Ordinary Least Squares (OLS) linear regression, the mean of residuals is always zero (or extremely close to zero due to rounding). 
- This happens because of how the regression line is mathematically chosen.


````python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([50,55,65,70,80])

model = LinearRegression().fit(X,y)

pred = model.predict(X)


residuals = y - pred

print("Residuals:", residuals)
print("Mean residual:", np.mean(residuals))
````

| $e_i$ or $\varepsilon_i$ | error / residual     |
